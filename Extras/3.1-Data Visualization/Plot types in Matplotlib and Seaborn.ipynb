```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
x = np.linspace(0, 50)
y = np.random.normal(0, 15, 50)
```

```python
# creating a dataframe
boston = pd.read_csv('./housing.csv', index_col = 0)
```

```python
sns.set_style('darkgrid')
```

```python
plt.figure(figsize = (10, 5))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('line graph')
```

```python
# Line plot

fig, ax = plt.subplots(2,2,figsize = (15,10))
ax[0,0].plot(x, y, label = 'xy')
ax[0,0].set(xlabel = 'x', ylabel= 'y',
           title = 'line graph')

ax[0,1].scatter(x, y, color = 'magenta', marker = 'x')
ax[0,1].set(xlabel = 'x', ylabel= 'y',
           title = 'scatter plot')
ax[1,0].hist(y, bins = 7)
ax[1,0].set(xlabel = 'y',
           title = 'histogram')

ax[1,1].hist2d(x, y)
ax[1,1].set(xlabel = 'x', ylabel= 'y',
           title = '2-d histogram')
```
