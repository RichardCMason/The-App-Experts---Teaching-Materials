```python
boston['LSTAT^2'] = boston['LSTAT'] ** 2
```

```python

boston['bias'] = 1
cols = boston.columns.difference(['Price'])
X = boston[cols].values
y = boston.Price.values
w = np.linalg.solve(np.dot(X.T, X),
                    np.dot(X.T, y))
y_hat = np.dot(X, w)
```

```python
def r2(y, y_hat):
    return 1 - np.sum((y-y_hat) ** 2) / np.sum((y - np.mean(y)) ** 2)
```

```python
r2(y, y_hat)
```

```python
plt.scatter(y, y-y_hat)
```
