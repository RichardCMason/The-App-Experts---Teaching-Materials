## 6. Calculate player "power".

---

Define a function that will calculate a player's "power". Player power is defined as the sum of the base statistics all of their pokemon.

$$
\text{player power } = \sum_{i = 1}^{n}\text{attack}_i + \text{defense}_i + \text{special attack}_i + \text{special defense}_i
$$

Where $i$ is an individual pokemon in a player's `player_pokemon`. ($\sum$ just means sum, so you're just adding up all the attributes listed above for all the pokemon in the player's `player_pokemon`).

Your function should:

*  Accept a `poke_players` dictionary and a `player_id` as arguments.
*  For the specified player_id, look up that player's pokemon.
*  Find and aggregate the attack and defense values for each of the player's pokemon.
*  Print "[player name]'s power is [player power].", where the player power is the sum of the base statistics for all of their pokemon.
*  Return the player's power value.

Check your work by displaying pokemon power for each of your players.

````python
def get_power(player_id, player_dict = poke_players):
    
    for pokemon in poke_players[player_id]['player_pokemon']:
        player_power = (poke_players[player_id]['player_pokemon'][pokemon]['hp'] + 
        poke_players[player_id]['player_pokemon'][pokemon]['attack'] +
        poke_players[player_id]['player_pokemon'][pokemon]['defense'] +           
        poke_players[player_id]['player_pokemon'][pokemon]['special_attack'] +
        poke_players[player_id]['player_pokemon'][pokemon]['special_defense'] +
        poke_players[player_id]['player_pokemon'][pokemon]['speed'])
    print(poke_players[player_id]['player_name'] + "'s power is " + str(player_power))
    return player_power

````

````python
def player_power(player_id,poke_players=poke_players):
    """Calculates player power by summing up attack, defence,special_attack and special defence of each of his pokemon
    and prints the name and power of the player.
    
    Parameters
    ----------
    player_id: `int`
        Unique identifier for each player.
    poke_players: `dict` 
        Player dictionary with default value `poke_players` containing the following fields:        
        - ``player_id`` : Id code unique to each player (`int`)
        - ``player`` : A dictionary defining the player (`dict`)
        
    Raises
    ------
    KeyError
        Raised if ``player_id`` us not in ``poke_players``.
    """
    if player_id in poke_players:
        player = poke_players[player_id]
        power = 0
        for poke_id,pokemon in player['player_pokemon'].items():
            power += pokemon['attack']+ pokemon['defense'] + pokemon['special_attack'] + pokemon['special_defense']
        print("{}'s power is {}".format(player['player_name'],power))
    else:
        raise KeyError("Invalid Player id")



````

````python
import pandas as pd

def filtered_pokedex(pokedex_data,filter_options,iqr=False):
    """ Filters the pokedex data according to the filter options provided.
    
    Parameters
    ---------- 
    pokedex_data : `dict`
        Pokedex data with the following fields:
        
        - PokedexNumber : An array of all the pokedex numbers associated with each pokemon (`np.array`).
        - Name : An array of all the names associated with each pokemon (`np.array`).
        - Type : An array of all the types asscociated with each pokemon (`np.array`).
        - Total : An array of total stats associated with each pokemon (`np.array`).
        - HP : An array of hitpoints associated with each pokemon (`np.array`).
        - Attack : An array of attack points associated with each pokemon (`np.array`).
        - Defense : An array of defence points associated with each pokemon (`np.array`).
        - SpecialAttack : An array of special attack points associated with each pokemon (`np.array`).
        - SpecialDefense : An array of special defence points associated with each pokemon (`np.array`).
        - Speed : An array of speed points associated with each pokemon (`np.array`).
        
    filter_options : `dict`
        Filter options according to which pokedex_data is to be filtered.
    iqr : `bool`
        A flag to determine whether the data is to be filtered by inter quantile range or not.
    
    Returns
    -------
    filtered_result : `list`
        The list of the filtered data from the pokedex_data where each element is the dictionary.
        
    Raises
    ------
    KeyError
        if an of keys in the ``filter_options`` is not in pokedex_data.
    
            
    """
    filtered_bool = np.array([True] * len(pokedex_data['Name']))
    for key,value in filter_options.items():
        if key not in pokedex_data:
            raise KeyError("Invalid key `{}`".format(key))
        
        if isinstance(value,str):
            filtered_bool = filtered_bool & (pokedex_data[key] == value)
        else:
            if iqr:
                to_compare = (value - 1.5  *stats.iqr(pokedex_data[key]), value + 1.5 * stats.iqr(pokedex_data[key]))
                filtered_bool = filtered_bool & ((pokedex_data[key] < to_compare[0]) | (pokedex_data[key] > to_compare[1]) )
            else:    
                filtered_bool = filtered_bool & (pokedex_data[key] >=value)
    filtered_data = {}
    for key,vals in pokedex_data.items():
        filtered_data[key] = vals[filtered_bool]
    df = pd.DataFrame(filtered_data)
    return [{column:item[column] for column in df.columns} for _,item in df.iterrows()]




````

````python
np.random.seed(1234)
def confidence_interval_90(pokedex_data=pokedex_data,n=50):
    """Calculates the 90 % confidence interval of the sample mean taken from the numeric columns of pokedex data.
    
    Parameters
    ----------
    pokedex_data : `dict`
        Pokedex data with `pokedex_data` as default containing the following fields:
        
        - PokedexNumber : An array of all the pokedex numbers associated with each pokemon (`np.array`).
        - Name : An array of all the names associated with each pokemon (`np.array`).
        - Type : An array of all the types asscociated with each pokemon (`np.array`).
        - Total : An array of total stats associated with each pokemon (`np.array`).
        - HP : An array of hitpoints associated with each pokemon (`np.array`).
        - Attack : An array of attack points associated with each pokemon (`np.array`).
        - Defense : An array of defence points associated with each pokemon (`np.array`).
        - SpecialAttack : An array of special attack points associated with each pokemon (`np.array`).
        - SpecialDefense : An array of special defence points associated with each pokemon (`np.array`).
        - Speed : An array of speed points associated with each pokemon (`np.array`).
    
    n : `int`
        Number of samples . `50` by default.
        
    Returns
    ------
    A dictionary with each numeric column of ``pokedex_data`` as a key and of the following form:
        key : (lower value of confidence interval,upper value of confidence interval) (`tuple`).
        
    """
    result_dict = {}
    for col in numeric_columns:
        data = np.random.choice(pokedex_data[col],n)
        std = np.std(pokedex_data[col])
        sample_mean = np.mean(data)
        result_dict[col] = (sample_mean - 1.645 * std/(np.sqrt(n)) , sample_mean + 1.645 *   std/(np.sqrt(n)))
    return result_dict
confidence_interval_90()
````
