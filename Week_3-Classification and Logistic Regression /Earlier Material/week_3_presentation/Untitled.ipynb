```python
from sklearn.model_selection import train_test_split
import pandas as pd
```

```python
data = pd.read_csv('./hfi_cc_2018.csv')
data.dropna(subset = ['hf_score'], inplace=True)
```

```python
train, test = train_test_split(data, train_size = .75, random_state = 0)
```

```python
train.info()
```

```python
test.info()
```

```python
train = train.reset_index(drop = True)
```

```python
train.head()
```

```python
train.drop(columns = ['hf_rank', 'hf_quartile'], axis = 1, inplace = True)
```

```python
train.dropna(subset = ['hf_score'], inplace=True)
```

```python
train.to_csv('train.csv')
```

```python
test = test.reset_index(drop = True)
```

```python
test.head()
```

```python
test.dropna(subset = ['hf_score'], inplace=True)
```

```python
solutions = pd.DataFrame({'Id':test.index,
                          'hf_score': test.hf_score})
```

```python
sample_submission = pd.DataFrame({'Id':test.index,
                          'hf_score': test.hf_score.mean()})
```

```python
sample_submission.to_csv('sample_submission.csv', index = False)
```

```python
solutions.to_csv('solutions.csv', index = False)
```

```python
test.drop(columns = ['hf_score', 'hf_rank', 'hf_quartile'], axis = 1, inplace=True)
```

```python
test.to_csv('test.csv')
```
