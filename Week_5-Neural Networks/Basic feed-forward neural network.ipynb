```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
D = 2 # number of features
K = 3 # of classes
N = K * 1000 # number
```

```python
# 3 groups of points
# 
X1 = np.random.randn(N//K, D) + np.array([2,2])
X2 = np.random.randn(N//K, D) + np.array([0,-2])
X3 = np.random.randn(N//K, D) + np.array([-2,2])

X = np.vstack((X1, X2, X3))
y = np.array([0]*(N//K) + [1]*(N//K) + [2]*(N//K))
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
def softmax(H):
    eH = np.exp(H)
    return eH/eH.sum(axis = 1, keepdims = True)
```

```python
def feed_forward(X, w1, b1, w2, b2):
    return softmax(np.tanh(X.dot(w1) + b1).dot(w2) + b2)
```

```python
M = 5

w1 = np.random.randn(D,M)
b1 = np.random.randn(1,M)
w2 = np.random.randn(M,K)
b2 = np.random.randn(1,K)


```

```python
P = feed_forward(X, w1, b1, w2, b2)
```
